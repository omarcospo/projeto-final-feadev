# Diferença entre PIB Real e Nominal

A conta feita é PIB Nominal - PIB Real

    - Se o Y nominal > Y real : inflação
    - Se o Y nominal < Y real : deflação

## Para analisar é interessante ter uma padronização dos dados

```python
def padronizar_percentual(df, colunas):
    for coluna in colunas:
        
        first_value = df[coluna].iloc[0]

        
        df[coluna] = (df[coluna] / first_value)

    return df

sp5002 = padronizar_percentual(sp5002, ['Fechamento'])
pib_us2 = padronizar_percentual(pib_us2, ['Monthly Nominal GDP Index', 'Monthly Real GDP Index'])

print(sp5002)
print(pib_us2)
```

```python
pib_us2.rename(columns={'Date': 'Data'}, inplace=True)
```

```python
merge_dados = pd.merge(pib_us2, sp5002, on='Data')

print(merge_dados)

print(merge_dados.dtypes)
```
